```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# mth partial sum of series solution
def u_trunc(x,t,m):
    u = 0
    for n in range(m):
        a_n = 2/np.pi*(-1)**n/(n+0.5)**2
        u = u + a_n*np.exp(-(n+0.5)**2*t)*np.sin((n+0.5)*x)
    return u

# plot truncated series for t=0,2 and m=1,2,5,10,100
N = 1e3
x_list = np.linspace(0,np.pi,int(N))
for t in [0, 2]:
    for m in [1, 2, 5, 10, 100]:
        u = [u_trunc(x,t,m) for x in x_list]
        plt.plot(x_list, u, label = "m = {}".format(m))
    plt.legend()
    plt.title("truncated series solution, t = {}".format(t))
    plt.xlabel("x")
    plt.ylabel("u".format(t))
    plt.savefig("hw9 q2 t={}".format(t))
    plt.show()
```

```python
t, m = 0, 100
u = np.array([u_trunc(x,t,m) for x in x_list])
diff = np.abs(u-x_list)
print("max diff for t = {} and m = {}: {}".format(t,m,np.max(diff)))
```
