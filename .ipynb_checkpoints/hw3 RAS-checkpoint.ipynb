```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
nx = 100
ny = 160
x = np.linspace(-5, 15, nx)
y = np.linspace(-10, 10, ny)
xg,yg = np.meshgrid(x, y)
z = xg + 1j*yg

# DIRK2
gamma = 1 - 1/np.sqrt(2)
f = (-gamma**2*z**2 + z)/(gamma**2*z**2 - 2*gamma*z + 1) + 1

# plot RAS
absf = (f.real)**2 + (f.imag)**2
plt.rcParams.update({'font.size': 22})
fig, ax = plt.subplots(figsize=(8,8))
plt.contourf(xg, yg, absf, np.arange(2))  
plt.title('DIRK2 RAS')
plt.xlabel("Re(z)")
plt.ylabel("Im(z)")
ax.set_aspect(1)
plt.grid(color='k', linestyle='--', linewidth=0.5)
plt.savefig('hw3 dirk2 RAS.png')
plt.show()
```

```python
# DIRK3 for gamma = 1+1/sqrt(2)
gamma = 1 + 1/np.sqrt(2)
f = 0.5*((1-4*gamma)*z**2+2*z)/(gamma**2*z**2-2*gamma*z+1)+1

# plot RAS
absf = (f.real)**2 + (f.imag)**2
plt.rcParams.update({'font.size': 15})
fig, ax = plt.subplots(figsize=(8,8))
plt.contourf(xg, yg, absf, np.arange(2))  
plt.title('DIRK3 RAS, gamma = 1+1/sqrt(2)')
plt.xlabel("Re(z)")
plt.ylabel("Im(z)")
ax.set_aspect(1)
plt.grid(color='k', linestyle='--', linewidth=0.5)
plt.savefig('hw3 dirk3 RAS 1.png')
plt.show()

# DIRK3 for gamma = 1-1/sqrt(2)
gamma = 1 - 1/np.sqrt(2)
f = 0.5*((1-4*gamma)*z**2+2*z)/(gamma**2*z**2-2*gamma*z+1)+1

# plot RAS
absf = (f.real)**2 + (f.imag)**2
plt.rcParams.update({'font.size': 15})
fig, ax = plt.subplots(figsize=(8,8))
plt.contourf(xg, yg, absf, np.arange(2))  
plt.title('DIRK3 RAS, gamma = 1-1/sqrt(2)')
plt.xlabel("Re(z)")
plt.ylabel("Im(z)")
ax.set_aspect(1)
plt.grid(color='k', linestyle='--', linewidth=0.5)
plt.savefig('hw3 dirk3 RAS 2.png')
plt.show()
```
